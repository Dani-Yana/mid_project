Adding district names instead of numbers to our dataframes. 

```python

def extract_district_num(data):
    ''' This function checks the length of the string
        and if lenght is 7 adds the first element of the 
        string to a list with district numbers. 
        If the lenght is 8 it adds the first two elements to the list'''
    district_num=[]
    for lor in data['lor']:
        if len(lor) == 7:
            district_num.append(lor[0])  
        elif len(lor) == 8:
            district_num.append(lor[0:2]) 

    return district_num
```

```python
district_numbers = extract_district_num(data1)
```

Determining if the theft happened during night or day with a funtion and applying on the dataset by creating a new columnn with the information:
- data from 4 columns : start date, end date, start time, end time
- night time definition in Germany 22:00 - 06:00
- check if start date is the same as end date and if start time and end time are between 6 and 22 -> day, else night
- check if end date is the next day and time is between 22 and 6 then -> night
- in any other case it can not be confirmed if the theft happened during day or night so it`s unknown

```python
def day_night(row):
    if row["end_date_incident"] == row["start_date_incident"]:
        if row["start_time_crime"]>=6 and row["end_time_crime"]<22:
            return "day"
        else:
            return "night"
    elif row["end_date_incident"]-row["start_date_incident"] == pd.Timedelta(days=1):
        if row["start_time_crime"]>=22 and row["end_time_crime"]<6:
            return "night"
        else:
            return "unknown"
    else:
        return "unknown"
```

```python
data1['Time_of_theft']=data1.apply(day_night,axis=1)
```
